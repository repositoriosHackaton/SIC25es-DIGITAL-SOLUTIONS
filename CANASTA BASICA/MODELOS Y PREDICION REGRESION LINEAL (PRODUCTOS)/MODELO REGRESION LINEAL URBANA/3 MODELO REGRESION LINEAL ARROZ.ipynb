```python
# Filtrar solo el arroz
df_Arroz = df[df["ARTICULO"] == "Arroz"].copy()


#df.drop(264, inplace=True)
df_Arroz = df_Arroz.drop(266, axis=0)

# Verificar el dataset filtrado

df_Arroz.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Arroz' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Arroz_melted = df_Arroz.melt(id_vars=["AÑO", "ARTICULO"], 
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", 
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Arroz_melted["MES"] = df_Arroz_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Arroz_melted["FECHA"] = pd.to_datetime(df_Arroz_melted["AÑO"].astype(str) + "-" + df_Arroz_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Arroz_melted = df_Arroz_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Arroz_melted.reset_index(drop=True, inplace=True)

df_Arroz_melted.index = df_Arroz_melted.index + 1


# Mostrar los primeros registros 
df_Arroz_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Arroz_melted["FECHA_ORD"] = df_Arroz_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Arroz_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Arroz_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime si no lo está
df_Arroz_melted["FECHA"] = pd.to_datetime(df_Arroz_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Arroz_melted["FECHA_ORD"] = df_Arroz_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Arroz_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Arroz_melted["PRECIO"].values       # Variable objetivo (precio)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12, 6))

# Datos históricos
plt.scatter(df_Arroz_melted["FECHA"], df_Arroz_melted["PRECIO"], 
            color="blue", s=10, alpha=0.8, label="Precio Real")

# Línea de regresión en datos históricos
plt.plot(df_Arroz_melted["FECHA"], y_pred, color="red", linestyle="dashed", 
         linewidth=2.5, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", s=15, 
            edgecolors="black", alpha=0.8, label="Predicciones 2025-2030")

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_min, y_max = df_Arroz_melted["PRECIO"].min() - 0.005, df_Arroz_melted["PRECIO"].max() + 0.02
plt.ylim(y_min, y_max)

# Formatear el eje Y
plt.yticks(np.arange(y_min, y_max, 0.005))

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio del Arroz (2025-2030)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")

precision_percentage = r2 * 100
print(f"Porcentaje de precisión IA: ", precision_percentage)
```

## __Análisis de la Evaluación del Modelo__

* Error Cuadrático Medio (MSE): 0.000059 → Muy pequeño, lo cual es positivo, ya que indica que las diferencias entre las predicciones y los valores reales son bajas.

* Raíz del Error Cuadrático Medio (RMSE): 0.007693 → En promedio, el modelo tiene un error de aproximadamente 0.77 centavos en la predicción del precio del arroz.

* Error Absoluto Medio (MAE): 0.005268 → Indica que, en promedio, el modelo se equivoca por 0.53 centavos por predicción.

* Coeficiente de Determinación (R²): 0.674217 → Indica que el 67.4% de la variabilidad en el precio del arroz es explicada por la fecha, lo que sugiere que hay otros factores que también influyen en los cambios de precio.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns

# Cálculo de los residuos
residuos = y - y_pred  

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_Arroz_melted["FECHA"], residuos, color="purple", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title(" Análisis de Residuos")
plt.grid()
plt.show()
```

## __Análisis de Residuos__

 Definición de residuos: Son la diferencia entre los valores reales y los valores predichos por el modelo. Si el modelo es preciso, los residuos deberían distribuirse aleatoriamente alrededor de cero.

   ###  __Patrón observado:__

    En los primeros años (2000-2010), los residuos muestran una tendencia a estar por encima de cero, lo que sugiere que el modelo subestima los precios en ese período.

    Entre 2010 y 2020, los residuos parecen disminuir, lo que indica una posible sobrestimación del modelo en ciertos momentos.
    
    Después de 2020, los residuos se acercan más a cero y parecen distribuirse de manera más uniforme.
