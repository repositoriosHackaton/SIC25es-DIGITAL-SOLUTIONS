# __ETAPA 2: FILTRAR LOS DATOS DE LAS FRUTAS__

```python
# Filtrar solo el Pan Frutas
df_Frutas_3 = df[df["ARTICULO"] == "Frutas 3_ /"].copy()


#df.drop(271, inplace=True)
df_Frutas_3 = df_Frutas_3.drop(271, axis=0)

# Verificar el dataset filtrado

df_Frutas_3.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Frutas_3' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Frutas_3_melted = df_Frutas_3.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Frutas_3_melted["MES"] = df_Frutas_3_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Frutas_3_melted["FECHA"] = pd.to_datetime(df_Frutas_3_melted["AÑO"].astype(str) + "-" + df_Frutas_3_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Frutas_3_melted = df_Frutas_3_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Frutas_3_melted.reset_index(drop=True, inplace=True)

df_Frutas_3_melted.index = df_Frutas_3_melted.index + 1


# Mostrar los primeros registros
df_Frutas_3_melted.head()
```

```python
#VERIFICACION CORRECTA DE LAS SERIES TEMPORALES
df_Frutas_3_melted
```

# __ETAPA 4: GRAFICAR LA EVOLUCION DEL PRECIO DE LAS FRUTAS__

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np

# Convertir la columna FECHA a datetime (si no lo está)
df_Frutas_3_melted["FECHA"] = pd.to_datetime(df_Frutas_3_melted["FECHA"])

# Crear la figura y el gráfico
plt.figure(figsize=(12,6))
plt.plot(df_Frutas_3_melted["FECHA"], df_Frutas_3_melted["PRECIO"],
         marker="o", linestyle="-", color="#FF5733", markersize=6,
         label="Precio Frutas")

# Personalizar el eje X (Años)
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

# Definir los valores del eje Y
y_ticks = np.arange(0.0600, 0.1901, 0.0050)
plt.yticks(y_ticks)

# Formatear las etiquetas del eje Y
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.4f'))

# Etiquetas y título
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Precio ($)", fontsize=12)
plt.title("Evolución del Precio de las Frutas", fontsize=14, fontweight="bold")

# Ajustes visuales
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 Analisi breve: Se observa una tendencia fluctuante en los precios de las frutas a lo largo del tiempo, con variaciones significativas.

Algunos picos podrían indicar aumentos estacionales o eventos de oferta y demanda en el mercado.

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Frutas_3_melted["FECHA_ORD"] = df_Frutas_3_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Frutas_3_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Frutas_3_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime
df_Frutas_3_melted["FECHA"] = pd.to_datetime(df_Frutas_3_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Frutas_3_melted["FECHA_ORD"] = df_Frutas_3_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Frutas_3_melted[["FECHA_ORD"]].values
y = df_Frutas_3_melted["PRECIO"].values

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Frutas_3_melted["FECHA"], df_Frutas_3_melted["PRECIO"],
            color="#1E88E5", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Frutas_3_melted["FECHA"], y_pred, color="#D81B60", linewidth=2,
         label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="#FFC107",
            label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

# Definir el rango del eje Y
y_ticks = np.arange(0.0600, 0.1901, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Precio ($)", fontsize=12)
plt.title("Evolución y Predicción del Precio de las Frutas (2025-2030)", fontsize=14, fontweight="bold")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

* Error Cuadrático Medio (MSE): 0.000073 → Un valor muy pequeño, lo que indica que el modelo tiene un buen ajuste.

+ Raíz del Error Cuadrático Medio (RMSE): 0.008570 → En promedio, el modelo tiene un error de aproximadamente 0.86 centavos en la predicción del precio de las frutas.

+ Error Absoluto Medio (MAE): 0.006733 → En promedio, el modelo se equivoca por 0.67 centavos por predicción, lo que muestra una alta precisión.

+ Coeficiente de Determinación (R²): 0.926873 → Indica que el 92.7% de la variabilidad en el precio de las frutas se explica por la fecha, lo que sugiere un excelente ajuste del modelo

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cálculo de los residuos
residuos = y - y_pred

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_Frutas_3_melted["FECHA"], residuos, color="#FF9800", alpha=0.7, label="Residuos")
plt.axhline(y=0, color="#37474F", linestyle="--", linewidth=1.5, label="Referencia (0)")

# Personalizar el gráfico
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Error de Predicción (Residuo)", fontsize=12)
plt.title("Análisis de Residuos", fontsize=14, fontweight="bold")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## __Análisis de Residuos__

  La gráfica de residuos nos permite evaluar qué tan bien el modelo de regresión lineal está ajustando los datos. En este caso, los puntos están distribuidos de manera relativamente uniforme alrededor de la línea horizontal en cero, lo que indica que el modelo no presenta un sesgo sistemático y que los errores de predicción están balanceados.

Sin embargo, si se observan patrones en los residuos por ejemplo, una tendencia creciente o decreciente.
