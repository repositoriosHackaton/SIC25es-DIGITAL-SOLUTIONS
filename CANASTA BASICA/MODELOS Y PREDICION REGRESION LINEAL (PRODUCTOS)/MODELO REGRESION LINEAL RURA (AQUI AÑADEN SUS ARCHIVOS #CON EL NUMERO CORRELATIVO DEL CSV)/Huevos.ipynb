# __ETAPA 2: FILTRAR LOS DATOS DEL HUEVO__

```python
# Filtrar solo el producto huevos
df_huevos = df[df["ARTICULO"] == "Huevos"].copy()


#df.drop(264, inplace=True)


# Verificar el dataset filtrado

df_huevos.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_huevos' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_huevos_melted = df_huevos.melt(id_vars=["AÑO", "ARTICULO"], 
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", 
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_huevos_melted["MES"] = df_huevos_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_huevos_melted["FECHA"] = pd.to_datetime(df_huevos_melted["AÑO"].astype(str) + "-" + df_huevos_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_huevos_melted = df_huevos_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_huevos_melted.reset_index(drop=True, inplace=True)

df_huevos_melted.index = df_huevos_melted.index + 1


# Mostrar los primeros registros 
df_huevos_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_huevos_melted["FECHA_ORD"] = df_huevos_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_huevos_melted[["FECHA_ORD"]]  # Matriz de características
y = df_huevos_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.dates as mdates

# Convertir la columna FECHA a datetime (si no lo está)
df_huevos_melted["FECHA"] = pd.to_datetime(df_huevos_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_huevos_melted["FECHA_ORD"] = df_huevos_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_huevos_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_huevos_melted["PRECIO"].values       # Variable objetivo (precio del pan)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_huevos_melted["FECHA"], df_huevos_melted["PRECIO"], color="blue", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_huevos_melted["FECHA"], y_pred, color="red", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y para que comience en 0.0000
plt.ylim(bottom=0.0000)  # Ajusta el límite inferior del eje Y a 0.0000

# Definir los valores del eje Y desde 0.0000 hacia el máximo
y_ticks = np.arange(0.0000, df_huevos_melted["PRECIO"].max() + 0.0050, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de los huevos (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación
precision_percentage = r2 * 100

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
print(f"Porcentaje de precisión IA: ", precision_percentage)
```

## __Análisis de la Evaluación del Modelo__

* Error Cuadrático Medio (MSE): 0.000038 = Muy pequeño, lo cual es bueno.

* Raíz del Error Cuadrático Medio (RMSE): 0.006152 = En promedio, el modelo tiene un error de aproximadamente 0.6 centavos en la predicción del precio del pan.

* Error Absoluto Medio (MAE): 0.004687 = Indica que, en promedio, el modelo se equivoca por 0.46 centavos por predicción.

* Coeficiente de Determinación (R²): 0.868891 = Indica que el 86% de la variabilidad en el precio de huevos se explica por la fecha.

* Porcentaje de precisión IA:  86.88906472161923 = Indica una alta precision en cuanto a las predicciones del modelo.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns

# Cálculo de los residuos
residuos = y - y_pred  

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_huevos_melted["FECHA"], residuos, color="purple", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title(" Análisis de Residuos")
plt.grid()
plt.show()
```

## __Análisis de Residuos__

    Los residuos (errores entre valores reales y predichos) están distribuidos en torno a la línea horizontal de referencia (0).

    Se observa un patrón no aleatorio en algunas secciones, lo cual sugiere que la regresión lineal no capta perfectamente todas las variaciones del precio de los huevos.

    La dispersión de los residuos parece aumentar en años recientes, lo que indica que el modelo puede tener más dificultades para ajustarse a cambios recientes en la dinámica del precio.

## __Histograma de Errores__ 

Este gráfico muestra cómo están distribuidos los errores del modelo.

```python
# Crear histograma de errores
plt.figure(figsize=(10,5))
sns.histplot(residuos, bins=20, kde=True, color="orange")

# Personalizar
plt.xlabel("Error de Predicción (Residuo)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Errores del Modelo")
plt.grid()
plt.show()
```

## __Distribución de Errores del Modelo__

    La mayoría de los errores están centrados alrededor de cero, lo que indica que el modelo tiene un buen desempeño general.

    Se observan algunos errores mayores a 0.015 en la parte derecha del gráfico, lo que implica que en algunos casos el modelo subestimó significativamente el precio real de los huevos.
